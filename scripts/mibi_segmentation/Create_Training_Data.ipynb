```python
from deepcell.utils.data_utils import make_training_data
NPZ_DIR = "/data/npz_data"
RESIZE = True
RESHAPE_SIZE = 256
DATA_DIR = "/data/contour_data/"
DATA_FILE = "Training_Freeze_1_Nuc"
import os
npz_name = "Training_Freeze_1_Nuc_256"
import numpy as np
import matplotlib.pyplot as plt
```

```python
make_training_data(
    #channel_names=['HH3.tif'],
    #channel_names=['BetaTubulin.tif', 'HH3.tif', "NaK ATPase.tif"],
    channel_names=['BetaTubulin.tif', 'HH3.tif', 'LaminAC.tif', 'NaK ATPase.tif', 'Vimentin.tif'],
    file_name_save=os.path.join(NPZ_DIR, npz_name),
    dimensionality=2,
    annotation_direc="annotated",
    annotation_name='Nuc_Mask_Label',
    reshape_size=RESHAPE_SIZE if RESIZE else None,
    training_direcs=None,
    raw_image_direc="raw",
    direc_name=os.path.join(DATA_DIR, DATA_FILE)
)
```

```python
# Load the training data from NPZ into a numpy array
training_data = np.load(os.path.join(NPZ_DIR, npz_name + ".npz"))

X, y = training_data["X"], training_data["y"]
```
