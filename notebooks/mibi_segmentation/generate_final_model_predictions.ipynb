```python
import os
import errno
import numpy as np 
import deepcell
import xarray as xr
from deepcell.applications import MultiplexSegmentation
from timeit import default_timer
```

```python
test_data = np.load('/data/npz_data/20201018_freeze/20201018_multiplex_final_seed_1_test_256x256.npz')
```

```python
X_test, y_test = test_data['X'], test_data['y']
```

```python
app = MultiplexSegmentation(use_pretrained_weights=False)
```

```python
# compartment
weights = "/data/models/20200619_production_updates/20201018_multiplex_final_seed_1_nuclear_deep_watershed.h5"
```

```python
app.model.load_weights(weights)
```

```python
output = app.predict(image=X_test, image_mpp=0.5, compartment='both', 
                     postprocess_kwargs_whole_cell={'maxima_threshold': 0.1, 'maxima_model_smooth': 0,
                                                   'interior_model_smooth': 2, 'interior_threshold': 0.23,
                                                   'small_objects_threshold': 15,
                                                   'fill_holes_threshold': 15,
                                                   'radius': 3}, 
                     postprocess_kwargs_nuclear={'maxima_threshold': 0.1, 'maxima_model_smooth': 0,
                                                   'interior_model_smooth': 2, 'interior_threshold': 0.23,
                                                   'small_objects_threshold': 15,
                                                   'fill_holes_threshold': 15,
                                                   'radius': 3,
                                                    'pixel_expansion': 3})
```

```python
output_no_expansion = app.predict(image=X_test, image_mpp=0.5, compartment='nuclear', 
                     
                     postprocess_kwargs_nuclear={'maxima_threshold': 0.1, 'maxima_model_smooth': 0,
                                                   'interior_model_smooth': 2, 'interior_threshold': 0.23,
                                                   'small_objects_threshold': 15,
                                                   'fill_holes_threshold': 15,
                                                   'radius': 3,
                                                    'pixel_expansion': 0})
```

```python
from deepcell.utils.plot_utils import create_rgb_image, make_outline_overlay
RGB_data = create_rgb_image(X_test[900:950], ['green', 'blue'])
```

```python
predicted_overlay = make_outline_overlay(RGB_data, output[900:950, :, :, :1])
```

```python
import matplotlib.pyplot as plt
cmap = plt.get_cmap('viridis')
cmap.set_bad('black')
index = 40
fig, axes = plt.subplots(figsize=(10,7))

axes.imshow(predicted_overlay[index, :, :, :], cmap=cmap)

axes.set_title('Predicted Overlay', fontsize=24)
```

```python
# for test data evaluation
save_dir = '/data/segmentation/20201022_test_split/'
np.savez_compressed(save_dir + '20201018_multiplex_final_seed_1_test_cell_prediction.npz', y=output[..., :1])
np.savez_compressed(save_dir + '20201018_multiplex_final_seed_1_test_nuc_expansion_prediction.npz', y=output[..., 1:])
np.savez_compressed(save_dir + '20201018_multiplex_final_seed_1_test_nuc_prediction.npz', y=output_no_expansion)
```

```python
# UMAP
from deepcell.utils.plot_utils import create_rgb_image, make_outline_overlay

arrays = ['COH_BC', 'COH_LN', 'CyCIF_Lung', 'CyCIF_Tonsil', 'Eliot', 'HIV', 'roshan']
for i in arrays:
    data_dir = '/data/segmentation/20200916_UMAP/'
    deepcell_input = xr.open_dataarray(data_dir + 'deepcell_input_{}.xr'.format(i))
    output = app.predict(image=deepcell_input.values, image_mpp=0.5, compartment='both', 
                     postprocess_kwargs_whole_cell={'maxima_threshold': 0.1, 'maxima_model_smooth': 0,
                                                   'interior_model_smooth': 2, 'interior_threshold': 0.23,
                                                   'small_objects_threshold': 15,
                                                   'fill_holes_threshold': 15,
                                                   'radius': 3},
                    postprocess_kwargs_nuclear={'maxima_threshold': 0.1, 'maxima_model_smooth': 0,
                                                   'interior_model_smooth': 2, 'interior_threshold': 0.23,
                                                   'small_objects_threshold': 15,
                                                   'fill_holes_threshold': 15,
                                                   'radius': 3,
                                                    'pixel_expansion': 0})
    
    segmentation_labels_cell = xr.DataArray(output,
                         coords=[deepcell_input.coords['fovs'], range(deepcell_input.shape[1]), 
                                 range(deepcell_input.shape[2]), 
                                 ['whole_cell', 'nuclear']],
                         dims=['fovs', 'rows', 'cols', 'compartments'])
    segmentation_labels_cell.to_netcdf(os.path.join(data_dir, "segmentation_labels_{}.xr".format(i)), 
                                       format="NETCDF3_64BIT")
```

```python
RGB_data = create_rgb_image(deepcell_input.values.astype('float'), ['green', 'blue'])
cell_overlay = make_outline_overlay(RGB_data, output[:10, :, :, :])
cell_overlay = make_outline_overlay(RGB_data, output[:10, :, :, 1:])
```
